### Labeling Experience Column for easy visualization

```python
def label_experience(min_exp):
    if min_exp <= 1:
        return 'Entry Level'
    elif min_exp <= 4:
        return 'Mid Level'
    else:
        return 'Senior Level'

new_df['Experience Band'] = new_df['Experience Min'].apply(label_experience)
```

```python
#lets drop the experience column now
new_df.drop(columns=['Experience'],inplace=True)
```

### working with salary column

```python
#only about 5% rows have salary
new_df['Salary Disclosed'] = new_df['Salary'].notnull()
new_df.fillna({'Salary':'Not Disclosed'}, inplace=True)
```

```python
mask = new_df['Salary'] != 'Not Disclosed'
new_df[['Salary Min', 'Salary Max']] = new_df['Salary'].apply(lambda s: pd.Series([float(n) for n in re.findall(r'\d+(?:\.\d+)?', str(s))[:2]] if 'LPA' in str(s) else [np.nan, np.nan]))
```

```python
new_df.head()
```

```python
def label_salary(s):
    if s < 3:
        return 'Low'
    elif s < 7:
        return 'Mid'
    elif s<=20:
        return 'High'
    elif pd.isna(s):
        return 'Not Disclosed'
    else:
        return 'Very High'

new_df['Salary Band'] = new_df['Salary Max'].apply(label_salary)
```

```python
#now lets remove the salary column
new_df.drop(columns=['Salary'],inplace=True)
```

```python
def extract_role(title):
    title = title.lower()
    keywords = ['software','engineer', 'developer', 'manager', 'executive', 'teacher', 'analyst', 'consultant','designer', 
                'architect', 'scientist','customer service', 'specialist', 'coordinator', 'administrator', 'director', 
                'supervisor', 'intern','sales']
    for word in keywords:
        if word in title:
            return word.capitalize()
    return 'Other'

new_df['Job Role'] = df['Title'].apply(extract_role)
```

### Geting the number of days job was posted

```python
def convert_to_days(row):
    unit = row['Posted Unit'].lower()
    value = row['Posted Value']
    if unit == 'day' or unit == 'days':
        return value
    elif unit == 'week' or unit == 'weeks':
        return value * 7
    elif unit == 'month' or unit == 'months':
        return value * 30
    else:
        return None

new_df['Posted Days Ago'] = new_df.apply(convert_to_days, axis=1)
```
