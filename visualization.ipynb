```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df=pd.read_csv('data/final_job_listings.csv')
```

```python
df.sample(5)
```

```python

fig = px.pie(df, names='Experience Band', title='Experience Band Distribution', hole=0.4)
fig.show()
```

```python
top_locations = df['Location'].value_counts().nlargest(10)
sns.barplot(x=top_locations.values, y=top_locations.index)
plt.title('Top 10 Job Locations')
plt.xlabel('Number of Jobs')
plt.ylabel('Location')
plt.show()
```

```python
heatmap_data = pd.crosstab(df['Job Role'], df['Experience Band'])
sns.heatmap(heatmap_data, annot=True, fmt='d', cmap='Blues')
plt.title('Job Role vs Experience Band')
plt.show()
```

```python
sns.histplot(data=df, x='Posted Days Ago', bins=30)
plt.title('Distribution of Job Posting Age')
plt.xlabel('Days Ago')
plt.ylabel('Number of Jobs')
plt.show()
```

```python
trend = df['Posted Days Ago'].value_counts().sort_index()
plt.plot(trend.index, trend.values, marker='o')
plt.title('Job Posting Trend Over Time')
plt.xlabel('Days Ago')
plt.ylabel('Number of Jobs Posted')
plt.gca().invert_xaxis()  # So "0 days ago" appears on the right
plt.show()
```
